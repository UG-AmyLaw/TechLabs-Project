```python
import pandas as pd
import numpy as np
import re
import string
```

```python
df_fake = pd.read_csv("Python_Techlabs/Fake.csv")
```

```python
df_true = pd.read_csv("Python_Techlabs/True.csv")
```

```python
df_fake
```

```python
df_fake = df_fake.drop(['date'], axis = 1)
```

```python
df_fake
```

```python
df_fake = df_fake.drop(['title'], axis = 1)
```

```python
df_fake = df_fake.drop(['subject'], axis = 1)
```

```python
df_fake['label'] = 0
```

```python
df_true = df_true.drop(['date'], axis = 1)
```

```python
df_true = df_true.drop(['title'], axis = 1)
```

```python
df_true = df_true.drop(['subject'], axis = 1)
```

```python
df_true['label'] = 1
```

```python
df_true
```

```python
frames = [df_fake, df_true]
```

```python
df = pd.concat(frames)
```

```python
df
```

```python
def word_drop(text):
    text = text.lower()
    text = re.sub('\[.*?\]', "", text)
    text = re.sub("\\W", " ", text)
    text = re.sub('<.*?>+', '', text)
    text = re.sub('[%s]'%re.escape(string.punctuation), '', text)
    text = re.sub('\n', '', text)
    text = re.sub('\w*\d\w*', '', text)
    return text
```

```python
df["text"] = df["text"].apply(word_drop)
```
